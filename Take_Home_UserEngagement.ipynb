```python
merged_data = pd.merge(df_user, df_engagement, left_on='object_id' , right_on='user_id')
merged_data
```

```python
#Separating date,week,year,month from timestamp
merged_data['date'] = merged_data['time_stamp'].astype('datetime64')
merged_data['week'] = merged_data['date'].dt.isocalendar().week
merged_data['year'] = merged_data['date'].dt.isocalendar().year
merged_data['month'] = merged_data['date'].dt.month

merged_data
```

```python
#Creating DF after grouping with the required data
grouped_data=merged_data.groupby(['user_id','year' ,'month', 'week','visited']).size()
#merged_data['weekly_visited']=grouped_data
new_df = grouped_data.to_frame(name = 'weekly_visited').reset_index()
new_df
```

```python
new_df = new_df.groupby('user_id')['weekly_visited'].max().to_frame(name = 'max_visit').reset_index()
new_df['adopted_user'] = (new_df['max_visit']>=3)
new_df
```

```python
df_final = pd.merge(df_user,new_df,how='outer',left_on='object_id',right_on='user_id')
df_final.head()
```

```python
df_corr=df_final
df_corr.drop(['object_id','creation_time','name','email','last_session_creation_time'],axis=1,inplace=True)
df_corr1=pd.get_dummies(df_corr)
df_corr1
```

```python

corrmat = df_corr1.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(10,10))
#plot heat map
g=sns.heatmap(df_corr1[top_corr_features].corr(),annot=True,cmap="coolwarm")
```

```python
from scipy import stats

# Function for calculating p-value matrix for multiple variables

def calculate_pvalues(df):
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            df_clean = df.loc[:,[r,c]].dropna()
            pvalues[r][c]=round(stats.spearmanr(df_clean.iloc[:,0], df_clean.iloc[:,1])[1], 4)
    return pvalues

calculate_pvalues(df_corr1)
```

```python
X = df_corr1.drop(labels=['adopted_user_True','user_id','max_visit'],axis=1)
y = df_corr1.adopted_user_True
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Applying SelectKBest class to score the 12 features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

# Concatenating two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']
print(featureScores.nlargest(12,'Score'))
```

From the above vissualizations, heatmap and p-values, we can conclude that the important features are - creation_source, org_id, invited_by_user_id
